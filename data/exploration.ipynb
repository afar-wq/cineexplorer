Valeurs manquantes et valeurs uniques par colonnes

```python
def describe(df):
    print("Shape:", df.shape)
    print("\nDtypes:\n", df.dtypes)
    print("\nMissing values:\n", df.isna().sum())
    print("\nEmpty strings:\n", (df == '').sum())
    print("\nUnique values:\n", df.nunique())

for name, df in [('movies', movies), ('genres', genres), ('persons', persons), ('principals', principals), ('titles', titles), ('professions', professions), ('knownformovies',knownformovies),('characters',characters),('directors',directors),('writers',writers),('ratings',ratings)]:
    print(f"--- {name} ---")
    describe(df)
    print("\n")
```

 Nombre moyen d’acteurs par film

```python
avg_actors = principals.groupby("('mid',)")["('pid',)"].count().mean()
print("nombre moyen d'acteur = ",avg_actors)
```

```python
genres["('mid',)"].isin(movies["('mid',)"]).all()
persons["('pid',)"].isin(persons["('pid',)"]).all()
principals["('mid',)"].isin(movies["('mid',)"]).all()
principals["('pid',)"].isin(persons["('pid',)"]).all()
titles["('mid',)"].isin(movies["('mid',)"]).all()
professions["('pid',)"].isin(persons["('pid',)"]).all()
knownformovies["('mid',)"].isin(movies["('mid',)"]).all()
knownformovies["('pid',)"].isin(persons["('pid',)"]).all()
```
